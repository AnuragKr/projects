```python
#Importing all necessary packages
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
def get_training_data():
  #File Path
  train_path = 'drive/MyDrive/Colab Notebooks/amazon_reviews_data/train.csv'
  #Training Data Set-Up
  df_train = pd.read_csv(train_path)
  #Adding columns as it is not present in file
  df_train.columns = ['polarity','title','text']
  X_train = df_train.drop(['polarity'],axis = 1)
  Y_train = df_train['polarity']
  return X_train,Y_train
```

```python
def get_testing_data():
  #File Path
  test_path = 'drive/MyDrive/Colab Notebooks/amazon_reviews_data/test.csv'
  #Test Data Set-Up
  df_test = pd.read_csv(test_path)
  #Adding columns as it is not present in file
  df_test.columns = ['polarity','title','text']
  X_test = df_test.drop(['polarity'],axis = 1)
  Y_test = df_test['polarity']
  return X_test,Y_test
```
